### Detect Sine Wave from an RF Signal

This example demonstrates how to detect a sine wave embedded in a noisy signal using the Fast Fourier Transform (FFT). We'll generate a synthetic signal, apply FFT, and then visualize the frequency spectrum.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the signal
fs = 1000  # Sampling frequency (Hz)
T = 1      # Signal duration (seconds)
N = int(fs * T) # Number of samples
t = np.linspace(0, T, N, endpoint=False) # Time vector

# Generate a clean sine wave
f_sine = 50 # Frequency of the sine wave (Hz)
amplitude_sine = 1
sine_wave = amplitude_sine * np.sin(2 * np.pi * f_sine * t)

# Generate some random noise
amplitude_noise = 0.5
noise = amplitude_noise * np.random.randn(N)

# Combine the sine wave and noise to simulate an RF signal
rf_signal = sine_wave + noise

# Plot the generated RF signal
plt.figure(figsize=(12, 6))
plt.plot(t, rf_signal)
plt.title('Simulated RF Signal (Sine Wave + Noise)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Now, we will apply the Fast Fourier Transform (FFT) to convert the signal from the time domain to the frequency domain. This allows us to see the constituent frequencies of the signal.

```python
# Perform FFT on the RF signal
# np.fft.fft computes the discrete Fourier Transform
yf = np.fft.fft(rf_signal)

# Compute the frequencies corresponding to the FFT output
# np.fft.fftfreq returns the DFT sample frequencies
x_freq = np.fft.fftfreq(N, 1 / fs)

# Take the magnitude of the FFT output and normalize it
# We only care about the positive frequencies, so we take the first half
y_magnitude = 2.0/N * np.abs(yf[0:N//2])
x_freq_positive = x_freq[0:N//2]

# Plot the magnitude spectrum
plt.figure(figsize=(12, 6))
plt.plot(x_freq_positive, y_magnitude)
plt.title('Magnitude Spectrum of the RF Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, fs / 2) # Limit x-axis to positive Nyquist frequency
plt.show()

# Identify the peak frequency
peak_frequency_index = np.argmax(y_magnitude)
peak_frequency = x_freq_positive[peak_frequency_index]

print(f"Detected peak frequency: {peak_frequency:.2f} Hz")
```

In the magnitude spectrum plot, you should observe a prominent peak at the frequency of the sine wave (in this case, 50 Hz), standing out against the noise floor. The `print` statement also explicitly shows the detected peak frequency.
